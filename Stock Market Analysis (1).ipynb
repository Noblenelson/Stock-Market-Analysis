```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
end= datetime.now()

start= datetime(end.year-1,end.month,end.day)
```

# Gather Stock Information

```python
#STOCK INFORMATION ON SELECT FIRMS IN THE RETAIL INDUSTRY BASED ON MARKET CAP

# Amazon
AMZN = data.DataReader('AMZN','yahoo',start,end)
# Apple
AAPL = data.DataReader('AAPL','yahoo',start,end)
# Alphabet inc
GOOG = data.DataReader('GOOG','yahoo', start,end)
# Tesla
TSLA = data.DataReader('TSLA', 'yahoo', start,end)
# Microsoft
MSFT = data.DataReader('MSFT','yahoo',start,end)
# Nvidia
NVDA = data.DataReader('NVDA','yahoo', start,end)
# Airbnb
ABNB = data.DataReader('ABNB','yahoo', start,end)
# Boeing co.
BA = data.DataReader('BA','yahoo', start,end)
# Johnson and Johnson
JNJ = data.DataReader('JNJ','yahoo', start,end)
```

```python
tickers = ['AMZN','AAPL','GOOG','TSLA','MSFT','NVDA', 'ABNB','BA','JNJ'] 
```

```python
consumer_stocks = pd.concat([AMZN,AAPL,GOOG,TSLA,MSFT,NVDA,ABNB,BA,JNJ], axis = 1, keys = tickers)
```

```python
consumer_stocks.columns.names = ['Consumer Ticker','Stock Info']
```

```python
consumer_stocks.head()
```

```python
consumer_stocks.to_csv('Consumer_Stock_Retail.csv')
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# Optional Plotly Method Imports
import plotly
import cufflinks as cf
cf.go_offline()
```

```python
AAPL['Adj Close'].plot(legend=True,figsize=(10,4))
```

```python
for tick in tickers:
    consumer_stocks[tick]['Adj Close'].plot(label = tick, figsize = (12,5))
plt.legend()
```

```python
AAPL['Volume'].plot(legend=True,figsize=(10,4))
```

```python
for tick in tickers:
    consumer_stocks[tick]['Volume'].plot(label = tick, figsize = (12,4))
plt.legend()
```

```python
# We'll use pct_change to find the percent change for each day
AAPL['Daily Return'] = AAPL['Adj Close'].pct_change()
# Then we'll plot the daily return percentage
AAPL['Daily Return'].plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
```

```python
sns.distplot(AAPL['Daily Return'].dropna(),bins=100,color='purple')
```

```python

# Create heatmap
sns.heatmap(consumer_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').corr(),annot = True,lw = 1)
```

```python

# Create cluter map
sns.clustermap(consumer_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').corr(), annot = True, lw = 1)
```
